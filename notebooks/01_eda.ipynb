# Phase 1 — Exploratory Data Analysis

UCI Forest Fires dataset: 517 rows, predict burned area (hectares).

```python
import sys
from pathlib import Path

ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
sys.path.insert(0, str(ROOT))

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

from src.data import load_config, load_raw_data

config = load_config(str(ROOT / "config.yaml"))
df = load_raw_data(config)
df.head()
```

```python
print(f"Rows: {len(df)}")
print(f"Zero-area fires: {(df['area'] == 0).sum()} ({100 * (df['area'] == 0).mean():.1f}%)")
print(f"Area skewness: {df['area'].skew():.2f}")
df.describe()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].hist(df["area"], bins=50, color="#FF6B35")
axes[0].set_title("Raw burned area (ha)")
axes[1].hist(np.log1p(df["area"]), bins=50, color="#457B9D")
axes[1].set_title("log1p(area)")
plt.tight_layout()
plt.show()
```

```python
num_cols = config["numerical_cols"] + [config["target_col"]]
sns.heatmap(df[num_cols].corr(), annot=True, fmt=".2f", cmap="RdYlBu_r", center=0)
plt.title("Correlation heatmap")
plt.tight_layout()
plt.show()
```
